# Building a Research Model in Seahorse

Click **Runtime → Run all**. In Colab, the first cell clones Seahorse, installs it, runs the research-model tutorial, and writes the benchmark artifacts.

## Setup

This cell is designed for one-click Colab usage. It also works locally from a cloned repository.

```python
import importlib.util
import os
import subprocess
import sys
from pathlib import Path

REPO_URL = "https://github.com/YahyaAalaila/seahorse.git"
NOTEBOOK_PATH = "docs/notebooks/04_building_a_research_model.ipynb"


def infer_github_ref(default="main"):
    try:
        from google.colab import output
        url = output.eval_js("window.location.href")
    except Exception:
        return default
    marker = "/blob/"
    if marker not in url or NOTEBOOK_PATH not in url:
        return default
    tail = url.split(marker, 1)[1]
    suffix = "/" + NOTEBOOK_PATH
    if suffix not in tail:
        return default
    return tail.rsplit(suffix, 1)[0]


REPO_REF = infer_github_ref()
REPO_DIR = Path("/content/seahorse") if Path("/content").exists() else Path.cwd()

if importlib.util.find_spec("unified_stpp") is None:
    if not (REPO_DIR / "pyproject.toml").exists():
        if REPO_DIR.exists() and REPO_DIR != Path.cwd():
            subprocess.run(["rm", "-rf", str(REPO_DIR)], check=True)
        subprocess.run([
            "git", "clone", "--branch", REPO_REF, "--single-branch",
            REPO_URL, str(REPO_DIR)
        ], check=True)
    os.chdir(REPO_DIR)
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-e", ".[docs]"], check=True)
else:
    candidate = Path.cwd()
    while candidate != candidate.parent and not (candidate / "pyproject.toml").exists():
        candidate = candidate.parent
    if (candidate / "pyproject.toml").exists():
        os.chdir(candidate)

print("Running from", Path.cwd())
print("GitHub ref:", REPO_REF)
```

## Run end-to-end tutorial

```python
import json
import runpy
import sys
from pathlib import Path

repo = Path.cwd().resolve()
sys.path.insert(0, str(repo / "examples"))
sys.argv = ["building_a_research_model.py", "--output", "runs/tutorials/building_a_research_model"]
summary = runpy.run_path(str(repo / "examples/building_a_research_model.py"), run_name="__main__")
```

## Inspect results

```python
from pathlib import Path
import json
out = Path("runs/tutorials/building_a_research_model")
print("Output directory:", out.resolve())
print(json.dumps(json.loads((out / "summary.json").read_text()), indent=2))
print("\nArtifacts:")
for p in sorted(out.rglob("*")):
    if p.is_file() and p.suffix.lower() in {".svg", ".html", ".json"}:
        print("-", p)
```
